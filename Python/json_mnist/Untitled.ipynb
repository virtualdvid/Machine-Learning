```python
new_file = []
for file in os.listdir(path):
    if file.endswith((".json")):
        label = file.split('.')[0]
        with open('{}/{}'.format(path, file)) as json_data:
            data = json.load(json_data)
            new_dict = {'label': label,
                        'image': data}
            new_file.append(new_dict)
            
with open('{}/digits.json'.format(path), 'w') as outfile:  
    json.dump(new_file, outfile)
```
